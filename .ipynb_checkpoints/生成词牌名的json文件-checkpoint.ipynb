```python
import pdfplumber
import re
import json

# ========================
# 1. ä» PDF ä¸­æå–å…¨éƒ¨æ–‡æœ¬
# ========================

def extract_pdf_text(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"
    return text

# æ›¿æ¢è¿™ä¸ªè·¯å¾„ä¸ºä½ è‡ªå·±çš„ PDF æ–‡ä»¶è·¯å¾„
PDF_PATH = "è¯ç‰Œåæ ¼å¼å¤§å…¨.pdf"

raw_text = extract_pdf_text(PDF_PATH)
print("âœ… PDF æ–‡æœ¬æå–å®Œæˆï¼Œå…± {} å­—".format(len(raw_text)))
```

```python
from docx import Document

def read_docx_text(file_path):
    doc = Document(file_path)
    return '\n'.join([para.text for para in doc.paragraphs])
```

```python
def convert_tone_to_symbols(pattern):
    """
    è½¬æ¢å¹³ä»„ä¸ºç¬¦å·ï¼š
    - å¹³     â†’ â—‹
    - ä»„     â†’ â—
    - (å¹³)   â†’ â—
    - (ä»„)   â†’ â—‰
    """
    pattern = pattern.replace('(å¹³)', 'â—')
    pattern = pattern.replace('(ä»„)', 'â—‰')
    pattern = pattern.replace('å¹³', 'â—‹')
    pattern = pattern.replace('ä»„', 'â—')
    return pattern
```

```python
import json
import re

def parse_multiple_ci_pai(text):
    """
    è§£æè¿ç»­æ’åˆ—çš„å¤šä¸ªè¯ç‰Œï¼Œä»¥ 'è¯ç‰Œå' ä¸ºæ¯ä¸ªè¯ç‰Œçš„å¼€å§‹
    """
    lines = [line.strip() for line in text.split('\n') if line.strip()]
    all_ci = {}
    i = 0
    n = len(lines)

    while i < n:
        if lines[i] == "è¯ç‰Œå" and i + 1 < n:
            # å¼€å§‹ä¸€ä¸ªæ–°çš„è¯ç‰Œ
            ci_name = lines[i + 1].strip("ã€Šã€‹")
            description = ""
            tone_pattern_line = ""
            i += 2  # ç§»åŠ¨åˆ°ä¸‹ä¸€è¡Œ

            # å‘åæŸ¥æ‰¾ æè¿° å’Œ è¯ç‰Œæ ¼å¼
            while i < n:
                if lines[i] == "æè¿°" and i + 1 < n:
                    description = lines[i + 1]
                    i += 2
                elif lines[i] == "è¯ç‰Œæ ¼å¼" and i + 1 < n:
                    tone_pattern_line = lines[i + 1]
                    i += 2
                    break  # æ ¼å¼ç»“æŸåè·³å‡ºï¼Œå‡†å¤‡ä¸‹ä¸€ä¸ªè¯ç‰Œ
                else:
                    i += 1

            # éªŒè¯æ˜¯å¦å®Œæ•´
            if not description or not tone_pattern_line:
                print(f"âš ï¸ è·³è¿‡è¯ç‰Œ '{ci_name}'ï¼šç¼ºå°‘æè¿°æˆ–æ ¼å¼")
                continue

            # åˆ†å¥ï¼ˆæŒ‰ä¸­æ–‡é€—å·ã€å¥å·ï¼‰
            sentences = re.split(r'[ï¼Œã€‚]', tone_pattern_line)
            sentences = [s.strip() for s in sentences if s.strip()]

            # è®¡ç®—æ¯å¥å­—æ•°ï¼ˆå»æ‰ (ä»„)ã€(å¹³)ã€å¹³ã€ä»„ï¼‰
            def count_chars(s):
                cleaned = re.sub(r'$$[^)]*$$', '', s)  # å»æ‰ (ä»„)
                cleaned = re.sub(r'[â—‹â—â—â—‰]', '', cleaned)  # å»æ‰ç¬¦å·
                return len(cleaned)

            sections = [{"chars": count_chars(s)} for s in sentences]

            # è½¬æ¢å¹³ä»„ä¸ºç¬¦å·
            tone_pattern = [convert_tone_to_symbols(s) for s in sentences]

            # å­˜å…¥ç»“æœ
            all_ci[ci_name] = {
                "description": description,
                "sections": sections,
                "tone_pattern": tone_pattern
            }
        else:
            i += 1

    return all_ci
```

```python
# âš ï¸ ä¿®æ”¹ä¸ºä½ çš„æ–‡ä»¶è·¯å¾„
file_path = 'è¯ç‰Œåæ ¼å¼å¤§å…¨.docx'  # æˆ– 'è¯ç‰Œ.txt'

try:
    # è¯»å–æ–‡ä»¶
    if file_path.endswith('.docx'):
        text = read_docx_text(file_path)
    else:
        with open(file_path, 'r', encoding='utf-8') as f:
            text = f.read()

    print("ğŸ“„ æ–‡ä»¶è¯»å–æˆåŠŸï¼Œå†…å®¹é¢„è§ˆï¼š\n")
    print(text[:500] + "...\n")

    # è§£ææ‰€æœ‰è¯ç‰Œ
    result = parse_multiple_ci_pai(text)

    if not result:
        print("âŒ æœªè§£æåˆ°ä»»ä½•è¯ç‰Œï¼Œè¯·æ£€æŸ¥æ ¼å¼ã€‚")
    else:
        print(f"âœ… æˆåŠŸè§£æ {len(result)} ä¸ªè¯ç‰Œï¼š{', '.join(result.keys())}")

        # ä¿å­˜ä¸º JSON
        output_file = 'cipai.json'
        with open(output_file, 'w', encoding='utf-8') as f:
            json.dump(result, f, ensure_ascii=False, indent=2)

        print(f"ğŸ“ å·²ä¿å­˜ä¸ºï¼š{output_file}\n")
        print("--- JSON é¢„è§ˆ ---")
        print(json.dumps(result, ensure_ascii=False, indent=2))

except Exception as e:
    print(f"âŒ é”™è¯¯ï¼š{e}")
    print("å¸¸è§é—®é¢˜ï¼š")
    print("  - æ–‡ä»¶è·¯å¾„é”™è¯¯")
    print("  - Word ä¸­æ®µè½æœªæ­£ç¡®åˆ†è¡Œ")
    print("  - ç¼ºå°‘â€˜è¯ç‰Œåâ€™ã€â€˜æè¿°â€™ã€â€˜è¯ç‰Œæ ¼å¼â€™ç­‰å…³é”®è¯")
```
